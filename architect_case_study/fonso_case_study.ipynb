```python
import pandas as pd
import networkx as nx
```

```python
#Import Data
G = nx.read_gexf("../graph.gexf")
data = pd.read_feather("../Data/ACM/AcmEPFL_paired_communes.feather")
data["Auteurs"]
```

```python
pr = nx.pagerank(G, alpha=0.9, weight='weight')
pr
```

```python
pr_data = pd.DataFrame(pr.items())
print(pr_data.describe())
print(pr["Concours pour une Ecole professionelle à Lausanne, jugement"])
print(pr_data[pr_data[1] == pr_data[1].max()].to_string())
```

```python
fonso_competitions = data[data["Auteurs"].str.contains("Fonso", na=False)]["Nom de l'objet"].to_list()
```

```python
fonso_comps_pr = {}
for key, item in pr.items():
    if key in fonso_competitions:
        fonso_comps_pr[key] = item
fonso_comps_pr = dict(sorted(fonso_comps_pr.items(), key=lambda item: item[1], reverse=True))
fonso_comps_pr
```
