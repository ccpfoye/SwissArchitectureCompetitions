```python
import networkx as nx
import pandas as pd
import numpy as np
from scipy.stats import zscore
```

```python
G = nx.read_gexf("../out/acm_authors_graph.gexf")
```

```python
centrality = nx.betweenness_centrality(G, normalized=True)
sorted_nodes = sorted(centrality.items(), key=lambda x: x[1], reverse=True)
```

```python
for node, score in sorted_nodes:
    print(f"{node}: {score:.4f}")
```

```python
pr = nx.pagerank(G, alpha=0.9, weight='weight')
pr
```

```python
pr_values = np.array(list(pr.values()))
z_scores = zscore(pr_values)
z_min = z_scores.min()
z_max = z_scores.max()
scaled_z_scores = (z_scores - z_min) / (z_max - z_min)
pagerank_z = dict(zip(pr.keys(), z_scores))
pagerank_z_scaled = dict(zip(pr.keys(), scaled_z_scores))
```

```python
pr_data = pd.DataFrame(pr.items())
print(pr_data.describe())
print(pr_data[pr_data[1] == pr_data[1].max()].to_string())
```

```python
pr_data.sort_values(1, ascending=False).head(10)
```

```python
pr_z_data = pd.DataFrame(pagerank_z.items())
print(pr_z_data.describe())
pr_z_data.sort_values(1, ascending=False).head(10)
```
