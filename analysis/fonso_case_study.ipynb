```python
import pandas as pd
import networkx as nx
import numpy as np
from scipy.stats import zscore
```

```python
#Import Data
G = nx.read_gexf("../out/acm_competitions_graph_domain.gexf")
data = pd.read_feather("../Data/ACM/AcmEPFL_paired_communes.feather")
data.columns
```

```python
pr = nx.pagerank(G, alpha=0.9, weight='weight')
pr
```

```python
pr_values = np.array(list(pr.values()))
z_scores = zscore(pr_values)
z_min = z_scores.min()
z_max = z_scores.max()
scaled_z_scores = (z_scores - z_min) / (z_max - z_min)
pagerank_z = dict(zip(pr.keys(), z_scores))
pagerank_z_scaled = dict(zip(pr.keys(), scaled_z_scores))
pagerank_z = dict(sorted(pagerank_z.items(), key=lambda item: item[1], reverse=True))
pagerank_z
```

```python
fonso_competitions = data[data["Auteurs"].str.contains("Fonso", na=False)]["Nom de l'objet"].to_list()
laverriere_competitions = data[data["Auteurs"].str.contains("Laverrière", na=False)]["Nom de l'objet"].to_list()
```

```python
data
```

```python
fonso_comps_pr = {}
for key, item in pr.items():
    try:
        year = str(data.loc[data['Nom de l\'objet'] == key, "Date de début de l'objet"].iloc[0])[:10]
        if key in fonso_competitions:
            fonso_comps_pr[f"{key} [{year}]"] = item
    except:
        pass
fonso_comps_pr = dict(sorted(fonso_comps_pr.items(), key=lambda item: item[1], reverse=True))
fonso_comps_pr
```

```python
laverriere_comps_pr = {}
for key, item in pr.items():
    try:
        year = str(data.loc[data['Nom de l\'objet'] == key, "Date de début de l'objet"].iloc[0])[:10]
        if key in laverriere_competitions:
            laverriere_comps_pr[f"{key} [{year}]"] = item
    except:
        pass
laverriere_comps_pr = dict(sorted(laverriere_comps_pr.items(), key=lambda item: item[1], reverse=True))
laverriere_comps_pr
```

```python
laverriere_comps_pr_z = {}
for key, item in pagerank_z.items():
    try:
        year = str(data.loc[data['Nom de l\'objet'] == key, "Date de début de l'objet"].iloc[0])[:10]
        if key in laverriere_competitions:
            laverriere_comps_pr_z[key] = {"year": year, "pr_z": float(item)}
    except:
        pass
laverriere_comps_pr_z = dict(sorted(laverriere_comps_pr_z.items(), key=lambda item: item[1]["pr_z"], reverse=True))
laverriere_comps_pr_z
```
